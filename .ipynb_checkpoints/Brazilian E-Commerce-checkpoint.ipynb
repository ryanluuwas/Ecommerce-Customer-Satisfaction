# Feature Engineering

We noted that there were many variables that were not significant with the data, as shown in the figure above, elements such as price, payment sequential, product name, had little effect on how they affected the review socre

However, we were also exposed to variables such as order item quantity which were influencing the review score. Although such fields were available in the data, they were far too less to build an accurate model to predict the review rating, hence we performed some analysis and were able to figure out additional information could leverage from the data to create new variables to help us with our model.

Firstly, we were able to understand that the data fields such as the order received date, delivery data as per customer, and delivery date as per courier, could be impacting the customers' reviews. TO understand if the number of days was one of the major factors influencing the customer review, we created new fields:

1. The estimated delivery time (the delivery time in which the product was promised to be delivered).
2. The customer delivery time (the actual time it took the product to reach the customer).
3. The courier delivery time (the delivery time as recorded by the courrier).

Furthermore, we are also curious about the impact of the shipping amounts on the order review score, and since the impact of both payment value and freight value were considerately low separately, we combined the two variables to a new variable - <b>Total order value.</b>

```python
## Replacing the null values with an 'Null' string
df_orders = df_orders.replace(np.nan, 'Null')

## Hardcoding the missing dates with the premissible dates from the other colums
df_orders['order_approved_at'] = np.where(df_orders['order_approved_at']=='Null', df_orders['order_purchase_timestamp'], df_orders['order_approved_at'])
df_orders['order_delivered_carrier_date'] = np.where(df_orders['order_delivered_carrier_date']=='Null', df_orders['order_estimated_delivery_date'], df_orders['order_delivered_carrier_date'])
df_orders['order_delivered_customer_date'] = np.where(df_orders['order_delivered_customer_date']=='Null', df_orders['order_estimated_delivery_date'], df_orders['order_delivered_customer_date'])

## Finding the difference in dates as per the delivery details 
df_orders['wd_estimated_delivery_time'] = np.busday_count(df_orders['order_approved_at'].values.astype('datetime64[D]')
                                                      ,df_orders['order_estimated_delivery_date'].values.astype('datetime64[D]'))

df_orders['wd_customer_delivery_time'] = np.busday_count(df_orders['order_delivered_customer_date'].values.astype('datetime64[D]')
                                                      ,df_orders['order_estimated_delivery_date'].values.astype('datetime64[D]'))

df_orders['wd_courier_delivery_time'] = np.busday_count(df_orders['order_delivered_carrier_date'].values.astype('datetime64[D]')
                                                      ,df_orders['order_estimated_delivery_date'].values.astype('datetime64[D]'))

## Finding out on which day of the week the order was placed
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['day_of_week'] = df_orders['order_purchase_timestamp'].dt.dayofweek
```

```python
# Merging data files with new engineered features
X = df_orders_2[['order_products_value', 'order_id', 'order_items_qty', 'order_sellers_qty']]

X1 = [X, df_orders, df_reviews, df_payments, df_items]

X1 = ft.reduce(lambda  left,right: pd.merge(left,right,on=['order_id'],
                                            how='outer'), X1)

X2 = pd.merge(X1,df_products,on=['product_id'],how='outer')

X3 = pd.merge(X2,df_sellers,on=['seller_id'],how='outer')

X4 = pd.merge(X3,df_customers,on=['customer_id'],how='outer')

df = X4.merge(df_product_category,on=['product_category_name'], how='outer').drop(['zip_code_prefix_y','zip_code_prefix_x'], axis=1)

df_pred = X2[X2['order_status'] == 'delivered']
```

```python
## Removing Nulls from the final dataset
df_pred[['product_category_name']] = df_pred[['product_category_name']].fillna("Not Available")
df_pred[['product_name_lenght']] = df_pred[['product_name_lenght']].fillna(value = 0)
df_pred[['product_description_lenght']] = df_pred[['product_description_lenght']].fillna("0")
df_pred[['product_photos_qty']] = df_pred[['product_photos_qty']].fillna(value = 0)

df_pred[['payment_value', 'payment_sequential' , 'payment_installments', 'payment_type', 'product_description_lenght']] = df_pred[['payment_value', 'payment_sequential', 'payment_installments', 'payment_type', 'product_description_lenght']].fillna(value = 0)

df_pred[['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']] = df_pred[['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']].fillna(value = 0)

df_pred[['order_products_value', 'order_items_qty', 'order_sellers_qty']] = df_pred[['order_products_value', 'order_items_qty', 'order_sellers_qty']].fillna(value = 0)

## Understanding if the customer delivery or the seller delivery is sooner

df_pred['courier_on_time'] = df_pred.apply(lambda row: 1 if row['order_delivered_customer_date'] <= row['order_estimated_delivery_date'] else 0, axis=1)
df_pred['seller_on_time'] = df_pred.apply(lambda row: 1 if row['order_delivered_carrier_date'] <= row['shipping_limit_date'] else 0, axis=1)

df_pred['product_description_lenght'] = df_pred['product_description_lenght'].astype(np.float64)

df_pred['total_order_value'] = df_pred.order_products_value + df_pred.freight_value

# df_pred['order_freight_ratio'] = df_pred.freight_value / df_pred.order_products_value
```

# Word Cloud and Text Data Vectorization Analysis

### What is a Word Cloud? 

* Generating a <b>Word Cloud</b> is a data visualization technique used for representing text data in which the size of each word indicates its frequency of importance. Significant textual data points can be highlighted using a word cloud. (GeeksforGeeks)

Since the data contains review score and actual reviews, we want these information available to us, to get an understanding and also corroborate the findings by analyzing the results we achieved from the predicting review scores and clustering review data based on their keywords.

The retrieved reviews, although in Portuguese, gave us a platform to compare our customer satisfaction results to a more realistic source that was within the data. 

Initially, we understood the total number of reviews that were present in the data and looked through their lengths to make sure there were no illicit comments or alphabets present which could deter our analysis. 

##### Removing empty observations 

```python
# Drop observations if review commment message or score is NA
df = df_reviews[['review_comment_message', 'review_score']].dropna()
```

##### Create new column 'comment length' 

```python
# Create new column 'comment length' based on the length of the review
df['comment_length'] = df['review_comment_message'].apply(len)

df.head()
```

```python
print('In the review dataframe, there are ' + 
      str(df.shape[0]) + 
      ' observations and ' + 
     str(df.shape[1]) +
     ' features.')
```

##### Creating a string for the Word Cloud to process 

```python
# Convert the review comments into a string to put into the word cloud
reviews = df[['review_comment_message']]

# Setting string to lowercase to make it easier to work with stop words.
string = (reviews.to_string()).lower()
```

##### Stop Word List

* <i>Used to remove filler words (i.e. I, me, my, myself, we, our, ours, ourselves, etc.) that may create too much noise.
* Because the reviews are in Portuguese, we will need to remove Portuguese filler words.

```python
s = stopwords.words('portuguese')
print(s)
```

##### Creating Word Cloud without 'Stop Words'

```python
# Feeding string into the wordcloud
wordcloud = WordCloud(stopwords=s, background_color="white").generate(string)

# Plot
plt.figure(figsize=(11,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

From the Google Translate, many of the reviews are great reviews. They mainly praised Olist for its great delivery schedule.

* <i> Of the total 119,151 observations, there are only 41,753 actual readable reviews, or 35.04% of the total observations. This is important to note when we peform our analysis. Because our analysis is based upon 35.04% of the total observations, <b>we cannot make a definitive conclusion on the whole dataset — only on those that did write a review</b>.

Then using the nltk.corpus library, we were able to perform <b>Term Frequency – Inverse Document Frequency</b>, known as TF-IDF, on the review data. 
* TF-IDF is an information retrieval technique that weighs a term’s frequency (TF) and its inverse document frequency (IDF). Each word or term has its respective TF and IDF score. The product of the TF and IDF scores of a term is called the TF-IDF weight of that term. To put it simply, the higher the TF-IDF score (weight), the rarer the term, and similarly, the lower the TF-IDF score, the more frequent the term.

```python
# Performing TF-IDF vectorization
vectorizer = TfidfVectorizer(max_features=100, ngram_range=(2, 4),
                             stop_words=s)
vectorizer.fit(reviews['review_comment_message'])
vect = vectorizer.transform(reviews['review_comment_message'])
```

We also performed <b>Principal Component Analysis</b>, known as PCA, on the vectorized result to reduce dimensionality by eliminating frequently occurring small filler words within the reviews. This also helps us reduce the processing time and with the visualizations. 

```python
pca = PCA(n_components=2, random_state=42)
pca.fit(vect.toarray())
vect_pca = pca.transform(vect.toarray())
plt.scatter(vect_pca[:, 0], vect_pca[:, 1])
```

Once we vectorized the data, we want to cluster the results into good and bad reviews. Due to the shape of the produced vectorized data graph, we were hesitant to use the K-means clustering. Because we start with random choice of clusters, the results produced by running the algorithm multiple times may differ. 

Instead, we decided on the agglomerative clustering approach,  a type of hierarchical clustering, where the results are reproducible. Agglomerative clustering uses a bottom-up approach, wherein each data point starts in its own cluster. These clusters are then joined, by taking the two most similar clusters together and merging them. You may notice that this is an CPU-intensive algorithm so it may take a while for it to run.

<b>Why we are clustering:</b>

Because the 'positive' words in the positive reviews are more frequent, thus having a higher TF-IDF score. Because the 'negative' words in the negative reviews are less frequent, they are likely to have a lower TF-IDF score. Naturally, we will have two clusters: one cluster will consists of words with low TF-IDF score (likely to be 'negative' words), and one cluster will consists of words with high TF-IDF score (likely to be 'positive' words.) 

```python
# Clustering the data into different parts

model = AgglomerativeClustering(n_clusters=2)
cluster = model.fit_predict(vect_pca)
plt.scatter(vect_pca[:, 0], vect_pca[:, 1], c=cluster);
```

```python
reviews['predicted_cluster'] = cluster
reviews.head()
```

```python
text = reviews[reviews['predicted_cluster'] == 0]['review_comment_message'].to_string()

wordcloud = WordCloud(background_color="white").generate(text)
plt.figure(figsize=(15, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off");
```

```python
text = reviews[reviews['predicted_cluster'] == 1]['review_comment_message'].to_string()

wordcloud = WordCloud(background_color="white").generate(text)
plt.figure(figsize=(15, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off");
```

From the two word clouds and with the help of Google Translate, we were able to figure out that the first set of word clusters were primarily filled with words related to good reviews and the second set of word clusters were dominant with the negative reviews.

```python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        return X[self.attribute_names]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([('selector', DataFrameSelector(num_attribs)),
                         ('std_scaller', StandardScaler())
                        ])
```

```python
train_orders_features_prepared = num_pipeline.fit_transform(train_orders_features)
train_orders_features_prepared

test_orders_features_prepared = num_pipeline.fit_transform(test_orders_features)
test_orders_features_prepared

some_data = train_orders_features.iloc[:8]
some_labels = train_orders_labels.iloc[:8]
some_data_prepared = num_pipeline.transform(some_data)
```

```python
linmodel = LinearRegression()
linmodel.fit(train_orders_features_prepared,train_orders_labels)

predictions = linmodel.predict(train_orders_features_prepared)
linear_mse = mean_squared_error(train_orders_labels, predictions)
linear_rmse = np.sqrt(linear_mse)
linear_rmse
```

```python
print('Predicted: {} \n Labels: {}'.format(list(linmodel.predict(some_data_prepared)), list(some_labels.values)))
```

# Conclusion

From our word cloud and customer satisfaction prediction, we learn that delivery time is an important factor in customer satisfaction. If the order delivered ahead of schedule, customers were more likely to be satisfied and leave a positive review.
* In other words, set the expectation low with long estimated delivery arrivals, but overperform by having the product arrive earlier. 

However, if the product exceeded the expected delivery date, customers were more likely to be unsatisfied. Alongside dealyed delivery schedule, wrong products delivered will result in negative reviews, as well. Therefore, we recommend Olist to evaluate the logistics process and resolve the issues why some orders were not successfully delivered in a timely manner.

High shipping fee is another factor that greatly influences customer satisfaction. Customers generally dislike having to pay for shipping fees. Because Olist is among the largest department stores and online e-commerce sites in Brazil, we recommend Olist to utilize their bargaining power and lower their shipping fees by negotiating with multiple carriers. 
* For example, customers would rather purchase a product that is 20 dollar product with free shipping than a 16 dollar product with a 4 dollar shipping fee. 

Another strategy Olist can implement to solve the high shipping fee complaints is to set a free shipping threshold because "nearly half (48 percent) of online shoppers admit to occasionally increasing the size of their orders to hit the free shipping threshold" (Mintel). Satisfied customers are likely to continue purchasing on Olist's platform, both third-party sellers and Olist benefit from the increased revenue gains.
